```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
default_figsize=(14, 8)

plt.figure(1, figsize=default_figsize)
x=np.linspace(0, 2*np.pi)

plt.subplot(4,2,1)
y=np.sin(x*16)
plt.plot(x,y,'r')

plt.subplot(4,4,3)
y=np.cos(x)+1
plt.plot(x,y,'orange')

plt.subplot(4,4,4)
y=np.arctan(x/6)
plt.plot(x,y,'yellow')

plt.subplot(4,1,2)
y=np.cos(x)
plt.plot(x,y,'g')

plt.subplot(2,4,5)
y=np.arctan(x)
plt.plot(x,y,'b')

plt.subplot(2,8,11)
y=np.tan(x*12)
plt.plot(x,y,'violet')

plt.subplot(2,8,12)
y=np.cosh(x)
plt.plot(x,y,'gray')

plt.subplot(2,2,4)
y=np.tan(x*12)
plt.plot(x,y,'black')
```
