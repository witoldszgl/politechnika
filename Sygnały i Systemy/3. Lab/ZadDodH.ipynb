```python
import math
import numpy as np
import matplotlib.pyplot as plt

a = input('Enter the length of the first side of the triangle: ')
b = input('Enter the length of the second side of the triangle: ')
angle = input('Enter the angle between those sides of the triangle (in degrees): ')

a = float(a)
b = float(b)
angle = float(angle)
#zamiana stopni na radiany
#print(np.pi)
radiusAngle = (angle*np.pi)/180

#z tw cosinusow: c^2 = a^2 + b^2 - 2ab*cos(angle)
# c = sqrt(a^2 + b^2 - 2ab*cos(angle))

c = math.sqrt(np.power(a,2) + np.power(b,2) - 2*a*b*np.cos(radiusAngle))
c = float(c)
print("Length of the third side: ", c)

#niech kazdy trojkat ma poczatek w punkcie (0,0)
default_figsize=(14, 8)
fig=plt.figure(1, figsize=default_figsize)

#obliczanie wspolrzednych b
#Xb^2 + Yb^2 = b^2
#sin(90-angle) = Yb/b  ->  Yb = b*sin(90-angle)
#Xb = sqrt(b^2 - (Yb*sin(90-angle)^2) )

if angle < 90:
    tempDegreeAngle = 90 - angle
    tempRadiusAngle = (tempDegreeAngle*np.pi)/180
    Yb = b*np.sin(tempRadiusAngle)
    Xb = np.sqrt(np.power(b,2) - np.power(Yb*tempRadiusAngle,2))
    x0, y0 = [0,0], [0,a]
    x1, y1 = [0,Xb], [0,Yb]
    x2, y2 = [0,Xb], [a,Yb]

if angle == 90:
    x0, y0 = [0,0], [0,a]
    x1, y1 = [0,b], [0,0]
    x2, y2 = [0,b], [a,0]

if angle > 90:
    tempDegreeAngle = 180 - angle
    tempRadiusAngle = (tempDegreeAngle*np.pi)/180
    Yb = b*np.sin(tempRadiusAngle)
    Xb = np.sqrt(np.power(b,2) - np.power(Yb*tempRadiusAngle,2))
    x0, y0 = [0,a], [0,0]
    x1, y1 = [0,-Xb], [0,Yb]
    x2, y2 = [-Xb,a], [Yb,0]

if angle > 180:
    print("Wrong value of degree!")

plt.plot(x0, y0, x1, y1, x2, y2)
plt.show()

#print("Yb i Xb: ")
#print(Yb, Xb)


```
